## Cargar la funciones uso agregado y luego merge shape, ejecutar las funciones secuencialmente!!

```python
def uso_agregado(data_in,data_out,clases):
    for keys, values in data_in.items():
        data_out[keys]= pd.merge(data_in[keys], clases, on ='USO', how = "left")
    return(data_out)
```

```python
dict_usos_agregados = {}
dic_usos_agregados_final = uso_agregado(dict_shapes,dict_usos_agregados, clases_agregadas)
```

```python
for keys, values in dic_usos_agregados_final.items():
    dic_usos_agregados_final[keys].to_file('C:/projectos/shapes/ag/'+ keys)
```

```python
import numpy as np
def merge_shape(data_in,data_out):
    for keys, values in data_in.items():
        data_in[keys] = data_in[keys].reset_index()
        USOS = data_in[keys].USO.unique()
        cod_usos = np.arange(1,len(USOS)+1)
        df_cod_usos = pd.DataFrame({'USO':USOS, "cod_usos":cod_usos})
        data_out[keys]= pd.merge(data_in[keys], df_cod_usos, on ='USO', how = "left")
    return(data_out)
```

```python
dict_merge = {}
dict_merge_final = merge_shape(dic_usos_agregados_final,dict_merge)
```
