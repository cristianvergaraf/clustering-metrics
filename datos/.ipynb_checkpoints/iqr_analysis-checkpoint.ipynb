# Analisis de datos atipicos IQR 

```python
import pandas as pd
```

```python
usos_wide_iqr = pd.read_csv('uso_wide_iqr.csv')
usos_wide = pd.read_csv('usos_wide.csv')
atipicos_iqr = pd.read_csv('atipicos_IQR.csv')
```

Ahora vamos a calcular el numero de casos de valores atipicos por ciudad

```python
usos_wide.head(5)
```

```python
usos_wide.columns
```

```python
usos_wide.pivot(values = ['COM', 'ED_SING', 'EQUIP_EDU', 'EQUIP_OTR',
       'EQUIP_SANI', 'HOS_REST', 'IND', 'IND_MX', 'OCIO_ESP', 'OFI', 'RES_PLU',
       'RES_PLU_MX', 'RES_UNI', 'RES_UNI_MX', 'SIN_EDIF'], columns = 'ciudades')
```

```python
pd.melt(usos_wide, id_vars = 'ciudades', value_vars = ['COM', 'ED_SING', 'EQUIP_EDU', 'EQUIP_OTR',
       'EQUIP_SANI', 'HOS_REST', 'IND', 'IND_MX', 'OCIO_ESP', 'OFI', 'RES_PLU',
       'RES_PLU_MX', 'RES_UNI', 'RES_UNI_MX', 'SIN_EDIF'])
```

```python
atipicos_iqr['ciudades'] = usos_wide.ciudades
atipicos_iqr = atipicos_iqr.iloc[:,1:19]
```

```python
atipicos_melt = pd.melt(atipicos_iqr, id_vars = ["ciudades"], value_vars = ['COM', 'ED_SING', 'EQUIP_EDU', 'EQUIP_OTR',
       'EQUIP_SANI', 'HOS_REST', 'IND', 'IND_MX', 'OCIO_ESP', 'OFI', 'RES_PLU',
       'RES_PLU_MX', 'RES_UNI', 'RES_UNI_MX', 'SIN_EDIF'])
```

```python
atipicos_melt.loc[:,['ciudades','value']]
```

Revisar el group by en esta pagina https://realpython.com/pandas-groupby/

```python
atipicos_ciudad = atipicos_melt.loc[:,['ciudades','value']].groupby('ciudades').value_counts()
```

```python
atipicos_ciudad = atipicos_melt.loc[:,['ciudades','value']].groupby('ciudades', as_index = False)['value'].value_counts()
```

```python
atipicos_ciudad
```

```python
atipicos_ciudad.to_csv('atipicos_ciudades.csv')
```

```python
atipicos_ciudades = atipicos_ciudad.query('value == "atipico"').sort_values('count', ascending = False)
```

```python
atipicos_ciudades
```

```python
atipicos_ciudades.to_csv('atipicos_ciudades.csv')
```
