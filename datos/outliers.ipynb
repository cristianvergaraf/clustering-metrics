```python
def find_iqr(x):
    return np.subtract(*np.percentile(x,[75,25]))
```

```python
def find_q1(x):
    return (np.percentile(x,[25]))
```

```python
def find_q3(x):
    return (np.percentile(x,[75]))
```

```python
q1 = usos_wide.iloc[:,1:18].apply(find_q1) 
q3 = usos_wide.iloc[:,1:18].apply(find_q3)
```

```python
iqr  = usos_wide.iloc[:,1:18].apply(find_iqr) 
```
