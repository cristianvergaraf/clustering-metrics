# Funcion para leer los nombres de los archivos txt. La funcion tiene como argumento la ruta del directorio donde se encuentran los archivos

```python
def leer_archivos(path):
    os.chdir(path)
    file_name = []
    for file in os.listdir():
        file_name.append(file)
    return (file_name)
```

### La siguiente funcion lee las métricas de forma cada documento y las junta en un nuevo dataframe.
### Los argumentos son:

 1.- Files: Un diccionario con el nombre de los archivos.

 2.- n = El número de lineas desde la parte final del documento donde se encuentran las metricas.
 
 3.- metricas = El nombre de las métricas que forman las columnas del nuevo dataset.


```python
def crear_df_metricas(files,n, metricas):
    dict_metricas = {}
    df_metricas = pd.DataFrame(columns= metricas)
    for file in files:
        n = n
        with open(file) as f:
            lines = f.readlines()[n:]
            for line in lines:
                x = line.split(" ", 1)
                dict_metricas[x[0]] = x[1][:-1]                 
                df_temp = pd.DataFrame(dict_metricas, index = [file])
            df_metricas = pd.concat([df_metricas,df_temp], axis = 0)   
    
    return df_metricas

            
        
```

```python
metricas_forma = ['PAFRAC', 'SHAPE_MN', 'SHAPE_MD', 'PARA_MN', 'PARA_MD', 'FRAC_MN', 'FRAC_MD', 'SQUARE_MN', 'SQUARE_MD'] 
```
