Vamos a definir una funcion que nos permita leer los archivos de cada comuna y a partir exportar un archivo csv.

```python
import pandas as pd
import os
import glob
import os
import string
```

```python
# La funcion debe contemplar definir la ruta donde se encuentra los archivos
```

```python
ruta_forma = r"C:/projectos/manifolds/ciudades/datos/metricas/forma/"
```

```python
def leer_archivos(path):
    os.chdir(path)
    file_name = []
    for file in os.listdir():
        file_name.append(file)
    return (file_name)
```

```python
acceso = "C://projectos//manifolds//ciudades//datos//metricas//areas"
```

```python
archivos_forma = leer_archivos(ruta_forma)
```

```python
archivos_forma
```

```python
def crear_df_metricas(files,n, metricas):
    dict_metricas = {}
    df_metricas = pd.DataFrame(columns= metricas)
    for file in files:
        n = n
        with open(file) as f:
            lines = f.readlines()[n:]
            for line in lines:
                x = line.split(" ", 1)
                dict_metricas[x[0]] = x[1][:-1]                 
                df_temp = pd.DataFrame(dict_metricas, index = [file])
            df_metricas = pd.concat([df_metricas,df_temp], axis = 0)   
    
    return df_metricas

            
        
```

```python
metricas = ['PAFRAC', 'SHAPE_MN', 'SHAPE_MD', 'PARA_MN', 'PARA_MD', 'FRAC_MN', 'FRAC_MD', 'SQUARE_MN', 'SQUARE_MD'] 
```

```python
ciudades_forma = crear_df_metricas(archivos_forma,-9,metricas)
```
